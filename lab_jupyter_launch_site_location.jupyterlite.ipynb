Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

```python
import folium
from folium.plugins import MarkerCluster

# Assuming spacex_df is your DataFrame
# Assuming spacex_df has the following columns: 'Launch Site', 'Lat', 'Long', 'class'


# Create MarkerCluster for grouping markers
marker_cluster = MarkerCluster().add_to(launch_map)

# Define marker colors based on the 'class' column
color_dict = {1: 'green', 0: 'red'}

# Create a dictionary to store cluster data
cluster_data = {}

#Add circles for each launch site
for site in launch_sites:
    launch_name = site["name"]
    launch_lat = site["lat"]
    launch_lon = site["lon"]

    # Create a Circle marker with popup label
    circle_marker = folium.Circle(
        location=(launch_lat, launch_lon),
        radius=100,
        color='#d35400',
        fill=True,
    ).add_child(folium.Popup(launch_name))

    # Add the Circle marker to the map
    circle_marker.add_to(launch_map)

# Create a Marker with custom label using DivIcon
    folium.Marker(
        location=(launch_lat, launch_lon),
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{launch_name}</b></div>'
        )
    ).add_to(launch_map)
    
# Iterate through each row in the DataFrame
for index, row in spacex_df.iterrows():
    launch_site = row['Launch Site']
    lat = row['Lat']
    lon = row['Long']
    marker_color = row['marker_color']
    
    # Customize Marker's icon based on success or failure
    icon = folium.Icon(color=marker_color)

    # Create a Marker object with its coordinate
    marker = folium.Marker(location=(lat, lon), popup=launch_site, icon=icon)

    # Add the Marker to the MarkerCluster
    marker.add_to(marker_cluster)

    # Update cluster data
    if launch_site not in cluster_data:
        cluster_data[launch_site] = {'lat': lat, 'lon': lon, 'color': marker_color}

# Save the map as an HTML file or display it
launch_map
# Alternatively, you can display the map directly in a Jupyter notebook
# launch_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
